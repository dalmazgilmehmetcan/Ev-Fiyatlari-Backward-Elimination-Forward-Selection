```python
# Gerekli kütüphanelerimizi tanımlıyoruz.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# Veri setimizi yüklüyoruz.
veriler = pd.read_csv('ev-fiyat-verisi.csv')
veriler.head()
```

```python
# Verimiz içerisindeki numerik ve kategorik değerleri ayırıyoruz.
sehir = veriler.iloc[:,0:1].values # kategorik
numeric_veri = veriler.iloc[:,1:].values

numeric_veri
```

```python
# Kategorik değer olan "şehir" değerine sırayla label ve one hot encoder uyguluyoruz.
le = LabelEncoder()
sehir[:,0] = le.fit_transform(sehir[:,0])


ohe = OneHotEncoder()
sehir=ohe.fit_transform(sehir).toarray()

sehir
```

```python
# Verimizi numpy dizisinden pandas dataframe'i haline getiriyoruz.
sehir_df = pd.DataFrame(data = sehir, index = range(22), columns=['ankara','bursa','istanbul'] )
numeric_veri_df =pd.DataFrame(data = numeric_veri, index = range(22), columns = ['oda_sayisi','salon_sayisi','ev_yasi','fiyat'])

#dataframe birlestirme islemi
veri_birlestirme=pd.concat([sehir_df,numeric_veri_df],axis=1)
veri_birlestirme
```

```python
# Fiyat tahmini yapacağımız için kolonları ayırıyoruz.

x = veri_birlestirme.iloc[:,0:6].values
y = veri_birlestirme.iloc[:,6:].values
x
```

```python
# Lineer regresyyon kullanrak modelimizi oluşturuyouz.
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)

y_pred = regressor.predict(x_test)
```

```python
# Modelimize forward selection uygulamak için verilerimizi uygun hale getiriyoruz.
x = pd.DataFrame(data = x, index = range(22), columns=['ankara','bursa','istanbul','oda_sayisi','salon_sayisi','ev_yasi'] )
print(type(x)) # forward selecetion için dataframe yapısında olmalıdır
print(type(y)) # forward selecetion için numpy dizisi haline olmalıdır
x
```

```python
# Forward Selection işlemimizi başlatıyoruz.
import statsmodels.api as sm
X=np.append(arr=np.ones((22,1)).astype(int), values=x, axis=1)
X_opt=x.iloc[:,[0,1,2,3,4,5]].values


regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
print(regressor_OLS.summary())
```

```python
# x1,x3 ve x4 değerleri aynı olan en düşük p value değerleri olduğundan hesaplamaya beraber eklenmektedir.
X=np.append(arr=np.ones((22,1)).astype(int), values=x, axis=1)
X_opt=x.iloc[:,[0,2,3]].values


regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
print(regressor_OLS.summary())
```

```python
# p value değerleri halen 0.05'ten küçük olduğundan bir sonraki en küçük p value değeri olan x6 sisteme eklenir.
X=np.append(arr=np.ones((22,1)).astype(int), values=x, axis=1)
X_opt=x.iloc[:,[0,2,3,5]].values


regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
print(regressor_OLS.summary())
```

```python
# p value değerleri halen 0.05'ten küçük olduğundan bir sonraki en küçük p value değeri olan x2 sisteme eklenir.
X=np.append(arr=np.ones((22,1)).astype(int), values=x, axis=1)
X_opt=x.iloc[:,[0,1,2,3,5]].values


regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
print(regressor_OLS.summary())
```

```python
# Son eklenen 1. index değerinden sonra p value değeri 0.05'i geçtiğinden dolayı 1.index sisteme eklenmemelidir.
# 1. index haricinde eklenmiş ve p value değeri 0.05'in altında olan 0,2,3,5 index değerleri ile model eğitimi gerçekleştirilmektedir.
x_train, x_test,y_train,y_test = train_test_split(x.iloc[:,[0,2,3,5]],y,test_size=0.33, random_state=0)

regressor = LinearRegression()
regressor.fit(x_train,y_train)

y_pred = regressor.predict(x_test)
```
